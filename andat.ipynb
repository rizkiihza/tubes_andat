```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("indicator-life_expectancy_at_birth.csv")
```

```python
df.dropna(inplace=True)

country_a = "Indonesia"
df_country_a = df[df['country'] == country_a].iloc[:, 146:]
list_country_a = list(df_country_a.values[0])
list_year = list(df_country_a)

# print(len(list_country_a), len(list_year))
sns.lineplot(x=list_year, y=list_country_a)
plt.show()
```

```python
df.dropna(inplace=True)
country_b = "Malaysia"
df_country_a = df[df['country'] == country_b].iloc[:, 146:]
list_country_a = list(df_country_a.values[0])
list_year = list(df_country_a)

# print(len(list_country_a), len(list_year))
sns.lineplot(x=list_year, y=list_country_a)
plt.show()
```

```python
df_country_a = pd.DataFrame({ "Year" : list_year, "Life Expectancy" : list_country_a})
filename_a = country_a + ".csv"
df_country_a.to_csv(filename_a)
```
